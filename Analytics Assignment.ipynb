# Sentiment Analysis to understand the brand image of swiggy on twitter 


```python
#To perform sentiment analysis i converted the words of a tweet into a 100- dimensional embedding vector using the pretrained
#Glove embedding vecor model.
glove_vectors = dict()
file = open('glove.6B.100d.txt', encoding = 'utf-8')
for line in file:
    values = line.split()
    word = values[0]
    vectors = np.asarray(values[1:])
    glove_vectors[word] = vectors
file.close()
```

```python
vec_dimension = 100
def get_vec(x):
    arr  = np.zeros(vec_dimension)
    text = str(x).split()
    for t in text:
        try:
            vec = glove_vectors.get(t).astype(float)
            arr = arr + vec
        except:
            pass
    arr = arr.reshape(1,-1)[0]
    return(arr/len(text))
```

```python

x = df.full_text.astype('string')
features = []
for tweet in x:
    feat = get_vec(tweet)
    features.append(feat)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
print(len(features),len(features[0]))
out = pca.fit_transform(features)
print(len(out),len(out[0]))
```

```python
#As the problem statement is of the unsupervised nature, i used clustering algorithms to find different the types of sentiments.
#1. Used K-Menas 
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, n_init= 5, max_iter=1000, random_state=420).fit(out)
```

```python
X = []
Y = []
Z = []
for item in out:
    X.append(item[0])
    Y.append(item[1])
    Z.append(item[2])

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(X,Y,Z, c = 'black')
plt.show()
```

```python
zeros_x = []
zeros_y = []
zeros_z = []
ones_y = []
ones_x = []
ones_z = []

for i in range(len(kmeans.labels_)):
    if kmeans.labels_[i]==0:
        zeros_x.append(out[i][0])
        zeros_y.append(out[i][1])
        zeros_z.append(out[i][2])
    else:
        ones_x.append(out[i][0])
        ones_y.append(out[i][1])
        ones_z.append(out[i][2])

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter(zeros_x, zeros_y, zeros_z, c = 'red', label = 'Cluster1')
ax.scatter(ones_x, ones_y, ones_z, c = 'blue', label = 'Cluster2')
plt.show()
```

#  priority order for social media people

```python
#To determine the priority of a given tweet,i define a priority score which takes in as factors like distance from the 
# centroid of the clusters and the no of friends, followers, tweets.
print(kmeans.cluster_centers_)
```

```python
center_1 = kmeans.cluster_centers_[0]
center_2 = kmeans.cluster_centers_[1]

total_followers = 0

followers = df.followers_count.astype('float32')
for items in followers:
    total_followers +=items
average_followers = total_followers/len(followers)
print(average_followers)

total_friends = 0

friends = df.friends_count.astype('float32')
for items in friends:
    total_friends +=items
average_friends = total_friends/len(friends)
print(average_friends)

total_retweets = 0

retweets = df.retweet_count.astype('float32')
for items in retweets:
    total_retweets +=items
average_retweets = total_retweets/len(retweets)
print(average_retweets)
```

```python
def distance(p1,p2):
    return (p1[0]-p2[0])**2+(p1[1]-p2[1])**2 + (p1[2]-p2[2])**2

####

def priority_score(coordinates,lmd,gamma,friends,followers,retweets):
    return (lmd/(0.00001+distance(center_1,coordinates)))+ (1/(0.00001+distance(center_2,coordinates))) + gamma*(friends/average_friends + followers/average_followers + retweets/average_retweets)
```

```python
print('Tweet 1-',x[3])

print('Friends, followers and retweet count -', df.friends_count[3],df.followers_count[3],df.retweet_count[3])
score_1 = priority_score(out[3],0.5,2,16,13,0)
print('Priority score for the tweet is =', score_1)
print('\n')
print('Tweet 2-',x[10])
print('Friends, followers and retweet count -', df.friends_count[10],df.followers_count[10],df.retweet_count[10])
score_2 = priority_score(out[10],0.5,2,129,102,0)
print('Priority score for the tweet is=', score_2)
```
